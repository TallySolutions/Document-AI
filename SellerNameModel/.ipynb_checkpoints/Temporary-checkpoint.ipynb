```python
st = StanfordNERTagger(stanford_classifier)
```

```python
imagesPath = '/home/aman/Documents/Tally/Git-Document-AI/Document-AI/BuyerNameModel/Dataset/SampleImages/8.jpeg'
image = preProcessImage(imagesPath)
data = pytesseract.image_to_data(image, output_type=Output.DATAFRAME)
data.dropna(inplace=True)
data.reset_index(inplace=True)
data.drop(columns=["index"], inplace=True)
data.head()
```

```python
texts = data["text"].to_numpy()
T = ""
for text in texts:
    T+=text + " "
tagged = st.tag(str(T).split())
names = []
for tag in tagged:
    if tag[1]=="ORGANIZATION":
        names.append(tag[0])
names
```

```python
indexes = []
for index in data.index:
    if data["text"][index] not in names:
        indexes.append(index)
D = data.drop(indexes)
D["x"] = D["left"] + D["width"]/2
D["y"] = D["top"] + D["height"]/2
```

```python
D
```

```python
X = D["x"].to_numpy()
Y = D["y"].to_numpy()
```

```python
plt.scatter(X, Y)
```

```python
from sklearn.cluster import KMeans
Kmeans = KMeans(n_clusters=2, n_init="auto", random_state=0)
```

```python
train = [[X[index], Y[index]] for index in range(len(X))]
```

```python
Kmeans.fit(train)
```

```python
for point in train:
    x = point[0]
    y = point[1]
    test = [[x, y]]
    pred = Kmeans.predict(test)
    if pred==0:
        plt.scatter(x, y, color="red")
    else:
        plt.scatter(x, y, color="black")
plt.show()
```

```python
D["cluster"] = 0
```

```python
for index in D.index:
    x = D["x"][index]
    y = D["y"][index]
    pred = Kmeans.predict([[x, y]])
    if pred==0:
        pass
    else:
        D["cluster"][index]=1
clear_output(wait=True)
D
```
