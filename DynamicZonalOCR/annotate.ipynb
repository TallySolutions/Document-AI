```python
import pandas as pd
import numpy as np
import os
from IPython.display import clear_output
import warnings
warnings.filterwarnings("ignore")
from sklearn.utils import shuffle
```

```python
df = pd.read_csv("Datasets/DBSCAN2.csv")
df.drop(columns=["Unnamed: 0"], inplace=True)
df["class"] = -1
df
```

```python
classes = {"id" : "invoiceDetails", "sd" : "sellerDetails", "bd" : "buyerDetails", "am" : "amount", "m" : "misc"}
```

```python
for index in df.index:
    print(index)
    print(df["text"][index])
    c = input("Enter class : ")
    df["class"][index] = classes[c]
    clear_output(wait=False)
```

```python
df
```

```python
for value in classes.values():
    print(value, " : ", len(df[df["class"]==value]))
```

```python
df = df[(df["class"]=="invoiceDetails") | (df["class"]=="sellerDetails") | (df["class"]=="buyerDetails") | (df["class"]=="amount")]
```
