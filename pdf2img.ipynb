```python
import os
import pdf2image
from pdf2image import convert_from_path
import cv2
import matplotlib.pyplot as plt
import PyPDF2
import PIL
from PIL import Image, ImageTk
import numpy as np
```

```python
pdfPath = os.path.join("samples")
print(pdfPath)
```

```python
imagesPath = os.path.join("samples", "images")
print(imagesPath)
```

```python
if not os.path.exists(imagesPath):
    if os.name=="posix":
        !mkdir -p {imagesPath}
    elif os.name=="nt":
        !mkdir {imagesPath}
```

```python
dir = os.listdir(pdfPath)
```

```python
len(dir)
```

```python
imgList=[]
for file in dir:
    split_tup = os.path.splitext(file)
    if split_tup[1] and split_tup[1]==".pdf":
        filePath = os.path.join(pdfPath, file)
        f = open(filePath, 'rb')
        readpdf = PyPDF2.PdfReader(f)
        totalpages = len(readpdf.pages)
        if totalpages==1:
            image = convert_from_path(filePath)
            imgList.append(image)
```

```python
len(imgList)
```

```python
for img in imgList:
    plt.imshow(img[0])
    plt.show()
```

```python
imageList=[]
for img in imgList:
    i=np.array(img[0])
    imageList.append(i)
```
